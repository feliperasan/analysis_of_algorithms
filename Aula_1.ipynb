# Projeto e Análise de Algoritmos

## Aula 1 - Exercicio 01

Qual é o menor valor de entrada n (considere n > 0) tal que um algoritmo cujo tempo de execução de 10n^2 é mais rápido do que um algoritmo cujo tempo de execução é 2^n?
Qual desses algoritmos você considera mais eficiente?

````python
n = [i for i in range(0, 11)]
````

````python
time_algoritmo_1 = [ 10 * element ** 2 for element in n]
````

````python
time_algoritmo_1
````

````python
time_algoritmo_2 = [2 ** element for element in n]
````

````python
time_algoritmo_2
````

````python
# O Algoritmo que tem como função um polinômio é mais eficiente para n > 10
# De acordo com o que foi visto em sala de aula os polinômios são mais eficientes
# que as funções exponenciais.
plt.plot(n, time_algoritmo_1, label='10*n^2')
plt.plot(n, time_algoritmo_2, label='2^n')
plt.xlabel('Quantidade Entrada')
plt.ylabel('Tempo de Execução')
plt.legend()
plt.show()
````

## Aula 1 - Exercicio 02

Cada um dos algoritmos abaixo recebe um inteiro positivo e
devolve outro inteiro positivo. Os dois algoritmos devolvem o
mesmo número se receberem o mesmo valor de entrada n?
Qual dos dois algoritmos é mais eficiente?

```shell
Soma-Quadrados-A (n)
1  x = 0 # 1
2  para j = 1 até n faça # n
3    x = x + j * j # 1
4  retorne (x) # 1

Soma-Quadrados-B (n)
1  x = n * (n + 1) * (2n + 1) # 1
2  x = x/6 # 1
3  retorne (x) # 1

# Soma-Quadrados-A = c1 + c2 x n + c3 + c4 = n x (c1 + c2 + c3 + c4) = n
# Soma-Quadrados-B = c1 + c2 + c3 = 1 (constante)
```

````python
n = [element for element in range(0, 20)]
````

````python
soma_quadrados_A = [element for element in n]
````

````python
soma_quadrados_A
````

````python
soma_quadrados_B = [1 for element in n]
````

````python
soma_quadrados_B
````

````python
x = [element for element in range(0, 20)]
y = [element for element in range(0, 20)]
plt.plot(n, soma_quadrados_A, label="Soma Quadrados A")
plt.xticks(x)
plt.plot(n, soma_quadrados_B, label="Soma Quadrados B")
plt.yticks(y)
plt.xlabel("Nº de Entrada")
plt.ylabel("Tempo de Execução")
plt.legend()
plt.show()
````
