## Aula 1 - Exercicio 03
Implemente em uma linguagem de programação a sua escolha o algoritmo abaixo, que verifica se o valor de entrada é ou não um número primo. Existe pior e melhor caso?
```shell
Algoritmo PRIMO (n)
1  j = 2  # 1
2  Enquanto j < n e mod(n,j) != 0 faça # 1 = t <= n - 1
3    j = j + 1    # 1
4  Se j = n # 1 ou 0
5    então escreva ("é primo") # 1
6  Senão # 1 ou 0
7    escreva ("não é primo") # 1

# Melhor caso ocorre se um número não for primo, quando o mod da divisão for igual a 0.

# Pior caso ocorre quando há um número primo, pois realizará o bloco enquanto (n - 1) vezes.

```

````python
def primo(n):
    j = int(2)
    while j < n and n % j != 0:
        j + j + 1

    if j == n:
        print("é primo")
    else:
        print("não é primo")


n = int(input("Enter a digit: "))

primo(n)
````

## Aula 1 - Exercicio 04
Implemente em uma linguagem de programação à sua escolha o algoritmo abaixo, que lista os números primos menores ou iguais ao valor de entrada. Existe pior e melhor caso?
```shell
Algoritmo LISTA_PRIMO (n)
1  x = 1  # 1
2  Para i = 2 até n faça  # n - 1
3    j = 2  # 1
4    Enquanto j < i e mod(i, j) != 0 faça #  n - 1
5      j = j + 1  # 1
6    Se j = i  # 1 ou 0
7      então A[x] = i  # 1
8      x = x + 1  # 1
9  Retorne A[x]  # 1

# O pior caso ocorre quando quando o laço enquanto é executado
# O melhor caso acontece quando j = i
```

````python
def lista_primo(n):
    x = 1
    i = 2
    a = []

    for i in range(n):
        j = 2
        while j < i and i % j != 0:
            j = j + 1

        if j == i:
            a.append(i)
            x = x + 1
    return a

n = int(input("Type a digit: "))
result = lista_primo(n)

print(lista_primo(n))
````

Compare assintoticamente o tempo de execução (apresente dados, tabelas e gráficos) dos algoritmos de ordenação:

Em seguida, responda aos seguintes questionamentos.

a. Qual desses algoritmos é o mais eficiente?

Resposta: O mais eficiente no melhor caso é o Insertion Sort, já no pior caso ambos os algoritmos com exceção do Insertion Sort, são equivalentes, ou seja, são O(n²).

b. A ordem inicial do vetor de entrada influencia o desempenho dos algoritmos?

Resposta: Sim, no Insertion e no Selection Sort a ordem tem diferença no tempo de execução do algoritmo, já no bubble sort não há essa influencia uma vez que o algoritmo varre todo o vetor independete de sua organização.

c. Qual desses algoritmos você usaria na sua aplicação?

Resposta: O Insertion Sort, pois tanto no pior quanto no melhor caso de acordo com a análise, demonstrou ser mais eficiente tendo seu melhor caso como O(n). Já em termos de Implementação o Selection Sort demonstra ser mais fácil de implementar.

### Insertion Sort

````python
%%time
# Complexidade do InsertionSort - PIOR CASO
# c1 x n + c2 + c3 + c4 + n + c5 + c6 + c7 + c8 = n²(c1 + c2 + ... + c8) = n² = O(n²)
# Complexidade - MELHOR CASO
# É O(n), caso o vetor esteja ordenado de forma crescente.
def insertion_sort(elements = []):
    for j in range(len(elements)):  # c1 -> n
        key = elements[j]  # c2 -> 1
        i = j - 1  # c3 -> 1
        while i >= 0 and elements[i] > key:  # c4 -> 1 <= t <= n
            elements[i + 1] = elements[i]  # c5 -> 1
            i -= 1  # c6 -> 1
        elements[i + 1] = key  # c7 -> 1
        print(f'Exec[{j}]: {elements}')
    return elements  # c8 -> 1

array = [66, 3, 36, 14, 51, 20, 28, 57, 23, 76, 96, 84, 95, 36, 82, 37, 29, 10]
result = insertion_sort(array)

# print(len(array))
print(f'Ordenação: {result}')
````

### Bubble Sort

````python
%%time
# Complexidade do Bubblesort - PIOR CASO
# c1 x n + c2 x n + c3 + c4 + c5 + c6 + c7 = n²(c1 + c2 + ... + c7) = n² = O(n²)
# Complexidade - MELHOR CASO
# Também é O(n²), pois ele precisa passar por todos os elementos para verificar se estão ordenados.
def bubble_sort(elements = []):
    for selected in range(len(elements)-1, 0, -1):  # c1 -> n
        for compared in range(selected):  # c2 -> n
            if elements[compared] > elements[compared + 1]:  # c3 -> 1
                temp = elements[compared]  # c4 -> 1
                elements[compared] = elements[compared + 1]  # c5 -> 1
                elements[compared + 1] = temp  # c6 -> 1
        print(f'Exec: {elements}')
    return elements  # c7 -> 1

array = [66, 3, 36, 14, 51, 20, 28, 57, 23, 76, 96, 84, 95, 36, 82, 37, 29, 10]
result = bubble_sort(array)

# print(len(array))
print(f'Ordenação: {result}')
````

### Selection Sort

````python
%%time
# Complexidade do Bubblesort - PIOR CASO
# c1 x n + c2 + c3 x n + c4 + c5 + c6 + c7 + c8 = n²(c1 + c2 + ... + c9) = n² = O(n²)
# Complexidade - MELHOR CASO
# Também é O(n²), porem caso o vetor esteja ordenado o algoritmo não entrará no condicional parar guardar o valor do index j
# nesse caso sua troca é constante.
def selection_sort(elements = []):
    for i in range(len(elements)-1, 0, -1):  # c1 -> n
        selected = i  # c2 -> 1
        for j in range(i + 1):  #  c3 -> n
            if elements[j] > elements[selected]:  # c4 -> 0 or 1
                selected = j  # c5 -> 1
        temp = elements[i]  # c6 -> 1
        elements[i] = elements[selected]  # c7 -> 1
        elements[selected] = temp  # c8 -> 1


        print(f'Selected[{elements[selected]}] - change[{elements[i]}]: {elements}')

    return elements  # c9 -> 1

array = [66, 3, 36, 14, 51, 20, 28, 57, 23, 76, 96, 84, 95, 36, 82, 37, 29, 2]
result = selection_sort(array)

# print(len(array))
print(f'Ordenação: {result}')
````
